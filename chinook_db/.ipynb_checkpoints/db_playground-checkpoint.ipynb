```python
#Lesson 12: Problem Set Deeper into SQL
#8. Quiz: Local Query - JOIN MediaType And Track

##  Here's a playground to help you get a little more comfortable using DB API
##  Nothing here is graded, you've done a ton of work so far and I think 
##  it'd be great to relax and play a bit!

##  Connect to the Chinook database and try a few of your own ideas.
##  This is also a great place to try experiments building queries 
##  for the next few quizzes!

import sqlite3

# Fetch records from either chinook.db
db = sqlite3.connect("chinook.db")
c = db.cursor()
QUERY = '''SELECT count(*) FROM Track, MediaType, Genre 
            WHERE Track.MediaTypeId = MediaType.MediaTypeId 
            AND Track.GenreId = Genre.GenreId
            AND Genre.Name = 'Pop' AND MediaType.Name = 'MPEG audio file';'''
c.execute(QUERY)
rows = c.fetchall()

'''Uncomment to see your query in python'''
#print "Row data:"
#print rows

'''Uncomment to print your query by row'''
print "your output:"
for row in rows:
    print "  ", row[0:]

'''Uncomment to see your query as a pandas dataframe.
This is similar to the output you've been seeing throughout this course
You can learn more about pandas dataframes in our Intro to Data Analysis course!'''

#import pandas as pd    
#df = pd.DataFrame(rows)
#print df

db.close()

```

```python
#Lesson 12: Problem Set Deeper into SQL
#9. Quiz: Local Query - Jazz track

##  Here's a playground to help you get a little more comfortable using DB API
##  Nothing here is graded, you've done a ton of work so far and I think 
##  it'd be great to relax and play a bit!

##  Connect to the Chinook database and try a few of your own ideas.
##  This is also a great place to try experiments building queries 
##  for the next few quizzes!

import sqlite3

# Fetch records from either chinook.db
db = sqlite3.connect("chinook.db")
c = db.cursor()
QUERY = '''SELECT DISTINCT Customer.FirstName FROM Customer, Invoice, InvoiceLine, Track, Genre
        WHERE Customer.CustomerId = Invoice.CustomerId 
        AND Invoice.InvoiceId = InvoiceLine.InvoiceId
        AND InvoiceLine.TrackId = Track.TrackId
        AND Track.GenreId = Genre.GenreId AND Genre.Name = 'Jazz';'''
c.execute(QUERY)
rows = c.fetchall()

'''Uncomment to see your query in python'''
#print "Row data:"
#print rows

'''Uncomment to print your query by row'''
#print "your output:"
#for row in rows:
#    print "  ", row[0:]

'''Uncomment to see your query as a pandas dataframe.
This is similar to the output you've been seeing throughout this course
You can learn more about pandas dataframes in our Intro to Data Analysis course!'''

import pandas as pd    
df = pd.DataFrame(rows)
print df
print '\n'
print len(df)

db.close()
```

```python
#Lesson 12: Problem Set Deeper into SQL
#10. Quiz: Local Query - Below Average Song Lengths
#which genre has the most songs of below average song length

##  Here's a playground to help you get a little more comfortable using DB API
##  Nothing here is graded, you've done a ton of work so far and I think 
##  it'd be great to relax and play a bit!

##  Connect to the Chinook database and try a few of your own ideas.
##  This is also a great place to try experiments building queries 
##  for the next few quizzes!

import sqlite3

# Fetch records from either chinook.db
db = sqlite3.connect("chinook.db")
c = db.cursor()
#QUERY = '''SELECT AVG(Milliseconds) FROM Track;'''
QUERY = '''SELECT Genre.Name, COUNT(*) FROM Track, Genre 
        WHERE Track.Milliseconds < (SELECT AVG(Track.Milliseconds) FROM Track) 
        AND Track.GenreId = Genre.GenreId
        GROUP BY Genre.GenreId'''
c.execute(QUERY)
rows = c.fetchall()

'''Uncomment to see your query in python'''
#print "Row data:"
#print rows

'''Uncomment to print your query by row'''
#print "your output:"
#for row in rows:
#    print "  ", row[0:]

'''Uncomment to see your query as a pandas dataframe.
This is similar to the output you've been seeing throughout this course
You can learn more about pandas dataframes in our Intro to Data Analysis course!'''

import pandas as pd    
df = pd.DataFrame(rows)
print df

db.close()
```
